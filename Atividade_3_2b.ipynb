### Deflexão em Viga Biapoiada com Carga Distribuída

Este notebook apresenta a solução para a deflexão de uma viga biapoiada sob um carregamento distribuído uniforme $q_0$.

#### Equação Diferencial da Deflexão

A equação diferencial que governa a deflexão de vigas é:

$$EI \frac{d^4 w}{dx^4} = q(x)$$

Para o caso de uma carga uniforme $q(x) = q_0$, a equação se torna:

$$EI \frac{d^4 w}{dx^4} = q_0$$

#### Integrais Sucessivas

A integração da equação diferencial leva às equações de momento, inclinação e deflexão, com constantes de integração:

* **Esforço Cortante, $V(x)$:**
    $$EI \frac{d^3 w}{dx^3} = V(x) = q_0 x + C_1$$

* **Momento Fletor, $M(x)$:**
    $$EI \frac{d^2 w}{dx^2} = M(x) = \frac{q_0 x^2}{2} + C_1 x + C_2$$

* **Inclinação, $\theta(x)$:**
    $$EI \frac{dw}{dx} = EI \theta(x) = \frac{q_0 x^3}{6} + \frac{C_1 x^2}{2} + C_2 x + C_3$$

* **Deflexão, $w(x)$:**
    $$EI w(x) = \frac{q_0 x^4}{24} + \frac{C_1 x^3}{6} + \frac{C_2 x^2}{2} + C_3 x + C_4$$

#### Condições de Contorno

Para uma viga biapoiada de comprimento $L$, as condições de contorno são:

* Deflexão nula nos apoios: $w(0) = 0$ e $w(L) = 0$.
* Momento fletor nulo nos apoios: $M(0) = 0$ e $M(L) = 0$.

Aplicando essas condições, determinamos as constantes de integração:

$$\fbox{$C_1 = -\frac{q_0 L}{2}$}$$

$$\fbox{$C_2 = 0$}$$

$$\fbox{$C_3 = \frac{q_0 L^3}{24}$}$$

$$\fbox{$C_4 = 0$}$$

#### Equações Finais

Substituindo as constantes, as equações finais para o esforço cortante, momento fletor, inclinação e deflexão são:

* **Esforço Cortante:**
    $$V(x) = q_0(x - \frac{L}{2})$$

* **Momento Fletor:**
    $$M(x) = \frac{q_0 x}{2}(x-L)$$

* **Inclinação:**
    $$EI \theta(x) = \frac{q_0}{24} (4x^3 - 6Lx^2 + L^3)$$

* **Deflexão:**
    $$y(x) = \frac{q_0 x}{24EI} (x^3 - 2Lx^2 + L^3)$$

```python


import numpy as np
import matplotlib.pyplot as plt
import deepxde as dde
import torch
dde.backend.backend_name = "pytorch"

# -----------------------
# Parâmetros físicos
# -----------------------
# Converter todas as unidades para o Sistema Internacional (SI)
# L = 6m
L = 6.0
# I = 22500 cm^4 = 22500 * (10^-2 m)^4 = 22500 * 10^-8 m^4 = 2.25e-4 m^4
I = 22500e-8
# E = 21000 kN/cm^2 = 21000 * 10^3 N / (10^-2 m)^2 = 21000 * 10^3 / 10^-4 N/m^2 = 21000 * 10^7 Pa = 2.1e11 Pa
E = 21000e7
# q = 2 kN/m = 2 * 10^3 N/m
q = 2e3
EI = E * I

# Solução exata para viga biapoiada com carga distribuída (MUDANÇA AQUI)
def func(x):
    # A fórmula da deflexão é w(x) = (q*x/(24*EI)) * (L^3 - 2*L*x^2 + x^3)
    # A carga é para baixo, então a deflexão é negativa.
    return - (q * x / (24 * EI)) * (L**3 - 2 * L * x**2 + x**3)

# Escalonamento de y
x_exato = np.linspace(0, L, 50).reshape(-1,1)
y_exato = func(x_exato)
y_scale = np.max(np.abs(y_exato))  # fator de escala para y

# -----------------------
# Geometria escalonada
# -----------------------
geom = dde.geometry.geometry_1d.Interval(0, 1)  # x em [0,1]

# -----------------------
# PDE escalonada (MUDANÇA AQUI)
# -----------------------
# A carga distribuída no problema é q, não P, e é para baixo (por isso o sinal negativo na PDE).
def ddy(x, y):
    return dde.grad.hessian(y, x)

def dddy(x, y):
    return dde.grad.jacobian(ddy(x, y), x)

def pde(x, y):
    # PDE original é d^4y/dx^4 = q/EI. Escalonada fica:
    # d^4y_bar/dx_bar^4 * y_scale/L^4 = q/EI
    # d^4y_bar/dx_bar^4 = (q*L^4)/(EI*y_scale)
    dy_xxxx = dde.grad.hessian(dde.grad.hessian(y, x), x)
    # Sinal negativo pois a carga q é para baixo.
    return dy_xxxx - (L**4 * (-q)) / (EI * y_scale)

# -----------------------
# Condições de contorno (MUDANÇA AQUI)
# -----------------------
def boundary_l(x, on_boundary):
    return on_boundary and dde.utils.isclose(x[0], 0)

def boundary_r(x, on_boundary):
    return on_boundary and dde.utils.isclose(x[0], 1)

# As condições de contorno para uma viga biapoiada são:
# Deflexão nula em x=0 e x=L
# Momento fletor nulo em x=0 e x=L
# M(x) = EI * d^2w/dx^2, então a segunda derivada é zero nos apoios.
bc1 = dde.icbc.DirichletBC(geom, lambda x: 0, boundary_l)  # y(0) = 0
bc2 = dde.icbc.DirichletBC(geom, lambda x: 0, boundary_r)  # y(L) = 0
bc3 = dde.icbc.OperatorBC(geom,lambda x, y, _: ddy(x, y), boundary_l) # y''(0) = 0
bc4 = dde.icbc.OperatorBC(geom,lambda x, y, _: ddy(x, y), boundary_r) # y''(L) = 0


# -----------------------
# Dataset
# -----------------------
data = dde.data.PDE(
    geom,
    pde,
    [bc1, bc2, bc3, bc4],
    num_domain=200,
    num_boundary=2,
)

# -----------------------
# Rede Neural
# -----------------------
layer_size = [1] + [40]*6 + [1]
activation = "tanh"
initializer = "Glorot uniform"
net = dde.nn.FNN(layer_size, activation, initializer)

model = dde.Model(data, net)
# O peso das condições de contorno pode precisar de ajuste, mas vamos começar com pesos iguais.
# Aqui estamos usando pesos para as 4 BCs, além da PDE.
model.compile("adam", lr=0.00001, loss_weights=[2, 100, 100, 100, 100])  #[1, 100, 10, 10, 100])
N_iter = 90000
step = 2000
losshistory, train_state = model.train(iterations=N_iter, display_every=step)

# -----------------------
# Predição
# -----------------------
x_pinn = np.linspace(0, 1, 50).reshape(-1,1)
y_pinn = model.predict(x_pinn)

# Converter para unidades físicas
x_phys = x_pinn * L
y_phys = y_pinn * y_scale

# -----------------------
# Plot PINN vs Solução Exata
# -----------------------
plt.figure(figsize=(8,5))
plt.scatter(x_exato, y_exato, label='Exata', color='blue', marker='o', s=50)
plt.plot(x_phys, y_phys, label='PINN', color='red', linestyle='--')
plt.xlabel("x [m]")
plt.ylabel("y(x) [m]")
plt.title("PINN vs Solução Exata (unidades físicas)")
plt.legend()
plt.grid()
plt.show()

# -----------------------
# Loss
# -----------------------
losses = np.array(losshistory.loss_train)
pde_loss_plot = losses[:,0]
icbc_loss_plot = np.sum(losses[:,1:], axis=1)
total_loss_plot = pde_loss_plot + icbc_loss_plot

plt.figure(figsize=(6,4))
epocas = np.linspace(0, N_iter, len(pde_loss_plot))
plt.plot(epocas, total_loss_plot, 'k', linewidth=2, label="Total")
plt.plot(epocas, pde_loss_plot, color='silver', linestyle='--', linewidth=2, label="DE loss")
plt.plot(epocas, icbc_loss_plot, color='r', linestyle='-.', linewidth=2, label="BC loss")
plt.xlabel('Épocas')
plt.ylabel('Função de Custo')
plt.yscale('log')
plt.title('Histórico da Função de Custo')
plt.legend()
plt.grid(True)
plt.show()

# -----------------------
# Deflexão máxima e rotação máxima (NOVA VERSÃO CORRIGIDA)
# -----------------------

# A deflexão máxima ocorre no centro da viga, x = L/2.
deflexao_max_analitica = func(np.array([[L/2]]))
print(f"Deflexão máxima (analítica): {deflexao_max_analitica[0][0]:.6f} m")

# Deflexão máxima PINN
max_deflection_pinn = np.min(y_phys)
print(f"Deflexão máxima (PINN): {max_deflection_pinn:.6f} m")

# A rotação máxima ocorre nos apoios (x=0 e x=L).
# A derivada da deflexão é y'(x) = q/(24*EI) * (4x^3 - 6Lx^2 + L^3)
# Rotação em x=0: y'(0) = q*L^3 / (24*EI)
theta_max_analitica = q * L**3 / (24 * EI)
print(f"Rotação máxima (analítica): {theta_max_analitica:.6f} rad")
```
