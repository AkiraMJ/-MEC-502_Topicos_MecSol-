##Gradiente de Função polinomial

````python
import torch
#HABILITANDO CÁLCULOS AUTOMÁTICOS DE DIFERENCIAÇÃO
x = torch.tensor(4.0,requires_grad=True)
y = torch.tensor(1.0, requires_grad=True)
z = torch.tensor(3.0, requires_grad=True)

#///FUNÇÃO POLINOMIAL:
f = 3*x**2*y+2*y**3+z**2+z

#CÁLCULO DE GRADIENTES (DERIVADAS PARCIAIS)
f.backward() #FUNÇÃO RESPONSÁVEL POR CALCULAR O GRADIENTE (BACKPROPAGATION)

print(f"df/dx = {x.grad.item()}")
print(f"df/dy = {y.grad.item()}")
print(f"df/dz = {z.grad.item()}")
````

# Análise do Código de Gradiente de Função Polinomial

O código calcula o gradiente (vetor de derivadas parciais) de uma função polinomial multivariada usando a diferenciação automática do PyTorch. Vamos analisar cada componente:

### 1. Definição das Variáveis
```python
x = torch.tensor(4.0, requires_grad=True)  # Ponto x=4.0
y = torch.tensor(1.0, requires_grad=True)  # Ponto y=1.0
z = torch.tensor(3.0, requires_grad=True)  # Ponto z=3.0
```
- `requires_grad=True` ativa o rastreamento de operações para cálculo de gradientes
- O gradiente será calculado no ponto (4, 1, 3)

### 2. Função Polinomial
```python
f = 3*x**2*y + 2*y**3 + z**2 + z
```
- Função tridimensional: f(x,y,z) = 3x²y + 2y³ + z² + z
- Combina termos quadráticos e cúbicos

### 3. Cálculo do Gradiente
```python
f.backward()  # Aciona a backpropagation para calcular gradientes
```
- Calcula automaticamente as derivadas parciais ∂f/∂x, ∂f/∂y e ∂f/∂z
- Armazena os resultados nos atributos `.grad` de cada variável

## Resultados e Verificação

### Cálculo Analítico
Para f(x,y,z) = 3x²y + 2y³ + z² + z:

1. **Derivada parcial em relação a x**:
   ∂f/∂x = 6xy  
   Em (4,1,3): 6*4*1 = **24**

2. **Derivada parcial em relação a y**:
   ∂f/∂y = 3x² + 6y²  
   Em (4,1,3): 3*16 + 6*1 = 48 + 6 = **54**

3. **Derivada parcial em relação a z**:
   ∂f/∂z = 2z + 1  
   Em (4,1,3): 2*3 + 1 = **7**

### Saída do Código
O código produzirá:
```
df/dx = 24.0
df/dy = 54.0
df/dz = 7.0
```

## Aplicações Práticas

O cálculo do gradiente é fundamental para:
1. **Otimização numérica**:
   - Métodos do gradiente descendente
   - Algoritmos de aprendizado de máquina

2. **Física computacional**:
   - Campos conservativos
   - Análise de potencial

3. **Engenharia**:
   - Sensibilidade de parâmetros
   - Análise de sistemas não-lineares

## Conclusão

Este código demonstra efetivamente:
- O cálculo automático de gradientes com PyTorch
- A aplicação em funções polinomiais multivariadas
- A precisão da diferenciação automática

Os resultados correspondem exatamente aos cálculos manuais, validando a implementação. O código pode ser facilmente estendido para funções mais complexas e maior dimensionalidade, sendo particularmente útil em problemas de otimização e aprendizado de máquina.

````python
#GRADIENTE DE FUNÇÃO TRIGONOMÉTRICA:
import torch
x = torch.tensor(10.0, requires_grad=True)
y = torch.tensor(2.0, requires_grad=True)
z = torch.tensor(3.0, requires_grad=True)

#///FUNÇÃO POLINOMIAL:
f = x**2 * torch.tan(y) + y * torch.sin(z)

f.backward()

print(f"df/dx = {x.grad.item()}")
print(f"df/dy = {y.grad.item()}")
print(f"df/dz = {z.grad.item()}")
````

# Análise do Código de Produto Escalar com PyTorch e NumPy

## Explicação do Código

O código demonstra o cálculo do produto escalar (também chamado de produto interno) usando duas bibliotecas diferentes: PyTorch e NumPy.

### Parte 1: Implementação com PyTorch
```python
import torch
x = torch.tensor([2.0, 3.0, 4.0])  # Cria tensor/vetor x
y = torch.tensor([5.0, 6.0, 7.0])  # Cria tensor/vetor y

```

### Parte 2: Implementação com NumPy
```python
import numpy as np
x = np.array([2.0, 3.0, 4.0])  # Cria array/vetor x
y = np.array([5.0, 6.0, 7.0])  # Cria array/vetor y

esc = np.dot(x, y)  # Calcula o produto escalar
print("Produto Escalar x,y", esc)  # Exibe o resultado
```

## Discussão dos Resultados

### Cálculo do Produto Escalar
O produto escalar entre dois vetores x = [x₁, x₂, x₃] e y = [y₁, y₂, y₃] é calculado como:
```
x · y = x₁*y₁ + x₂*y₂ + x₃*y₃
```

Para os vetores dados:
```
x = [2.0, 3.0, 4.0]
y = [5.0, 6.0, 7.0]
```

O cálculo manual seria:
```
2.0*5.0 + 3.0*6.0 + 4.0*7.0 = 10 + 18 + 28 = 56
```

### Comparação entre PyTorch e NumPy

1. **Resultados**:
   - Ambos os métodos (PyTorch e NumPy) retornarão exatamente o mesmo valor: 56.0
   - A pequena diferença é que PyTorch retorna um tensor, enquanto NumPy retorna um array numpy

2. **Performance**:
   - Para operações simples como esta, o desempenho é praticamente idêntico
   - NumPy pode ser ligeiramente mais rápido para operações puramente numéricas
   - PyTorch tem vantagem quando usada com GPUs para operações mais complexas

3. **Casos de Uso**:
   - **PyTorch**: Ideal para computação que requer diferenciação automática (como em redes neurais)
   - **NumPy**: Melhor para computação científica geral sem necessidade de gradientes

## Discussão dos Resultados

### Cálculo do Produto Vetorial
Para vetores tridimensionais x = [x₁, x₂, x₃] e y = [y₁, y₂, y₃], o produto vetorial é calculado como:

```
x × y = [x₂y₃ - x₃y₂, x₃y₁ - x₁y₃, x₁y₂ - x₂y₁]
```

Aplicando aos vetores dados:
```
x = [1.0, 2.0, 3.0]
y = [4.0, 5.0, 6.0]
```

O cálculo manual seria:
```
[2*6 - 3*5, 3*4 - 1*6, 1*5 - 2*4] = [12-15, 12-6, 5-8] = [-3, 6, -3]
```

## Discussão dos Resultados

### Cálculo Manual
Para uma matriz A (m×n) e vetor v (n×1), o produto Av resulta em um vetor (m×1) onde cada elemento i é calculado como:

Av[i] = Σ (A[i,j] * v[j]) para j de 1 a n

No exemplo:
```
x = [[1,2],   y = [7,8]
    [3,4],
    [5,6]]
```

Cálculo manual:
```
[1*7 + 2*8] = [7+16] = [23]
[3*7 + 4*8]   [21+32]   [53]
[5*7 + 6*8]   [35+48]   [83]
```

# Análise do Código de Transposição de Matrizes com PyTorch e NumPy

```python
import torch

x = torch.tensor([[1,2,3],[4,5,6]])
```

A transposição de uma matriz inverte suas dimensões, transformando linhas em colunas e vice-versa. Para uma matriz m×n, a transposta será n×m.

Matriz original x (2×3):
```

 [4, 5, 6]]
```

Matriz transposta xt (3×2):
```

```

### Comparação entre PyTorch e NumPy

   - PyTorch: `torch.transpose(tensor, dim0, dim1)`
   - NumPy: `np.transpose(array)` ou `array.T`
   - Ambas permitem a forma abreviada `.T` para transposição simples

2. **Resultados**:
   - Produzem exatamente a mesma operação matemática
   - PyTorch mantém a capacidade de cálculo de gradientes
   - NumPy oferece operações otimizadas para CPU

3. **Performance**:
   - Para matrizes pequenas, diferença insignificante

````python
#CÁLCULO DE DIVERGENTE
import torch
#DIVERGENTE DO CAMPO VETORIAL
#DECLARAÇÃO DAS VARIÁVEIS
x=torch.tensor(1.0,requires_grad=True)
y=torch.tensor(2.0,requires_grad=True)
z=torch.tensor(1.0,requires_grad=True)
#CAMPO VETORIAL
F=torch.stack([x**2+y+2*z, x+y**2+z, x*2+y+z**2])

div=0.0
#ESTRUTURA DE LOOP:
#ATRIBUI INDICE AS VARiÁVEIS(PODE SE ALTERAR A ORDEM DE X, Y E Z)
for i, var in enumerate([y,x,z]):
# RESETA OS VALORES DOS GRADIENTES
    x.grad=y.grad=z.grad=None
#CALCULA OS GRADIENTES DE CADA COMPONENTE SEGUNDO I'ÉSIMO ELEMENTO
    F[i].backward(retain_graph=True)
#GUARDA VALOR NUMÉRICO DO GRADIENTE
    grad_i = var.grad.item()
#SOMA OS VALORES NUMÈRICOS DO GRADIENTE
    div += grad_i

print("Divergente de F(y,x,z):",div)
````

````python
#CÁLCULO DE MATRIZ JACOBIANA
import torch
#DEFINIçâO DAS VARIÁVEIS
x=torch.tensor(1.0,requires_grad=True)
y=torch.tensor(2.0,requires_grad=True)
def F(x,y):
#EMPILHA OS VETORES EM LINHAS
    return torch.stack([x**2+y**2, torch.cos(x*y)])
#CAMPO VETORIAL
F=F(x,y)
#RESPOSTA PARA X E Y SUBSTITUIDOS NO PAR DE FUNÇÔES DECLARADAS EM torch.stack
#print("F",F)
#RESPOSTA DA FUNÇÂO TRIGONOMÈTRICA SAI EM RADIANOS
# ARMAZENA VETORES LINHA DA JACOBIANA
jacobian=[]

for i in range(len(F)):
# RESETA OS VALORES DOS GRADIENTES
    x.grad = y.grad = None
#CALCULA OS GRADIENTES DE CADA COMPONENTE
    F[i].backward(retain_graph=True)
#ADICIONA OS GRADIENTES EM jacobian[]
    jacobian.append([x.grad.item(), y.grad.item()])
#ARMAZENA A JACOBIANA EM FORMA DE TENSOR
J = torch.tensor(jacobian)
#RETORNA A JACOBIANA EM FORMA DE TENSOR
print("Matriz Jacobiana")
print(J)
````

````python
#CÁLCULO DE ROTACIONAL
import torch

# Define as variáveis simbólicas com requires_grad=True
x=torch.tensor(1.0,requires_grad=True)
y=torch.tensor(2.0,requires_grad=True)
z=torch.tensor(3.0,requires_grad=True)

#DEFINIDO O CAMPO VETORIAL
F1=y*z
F2=x*z
F3=x*y
#EMPILHAMENTO DOS VETORES
F=torch.stack([F1,F2,F3])
# CALCULA AS DERIVADAS PARCIAIS PARA O ROTACIONAL
# COM RETAIN_GRAPH=TRUE PARA PODER USAR A MESMA COMPUTAÇÃO MAIS DE UMA VEZ
rotx=torch.autograd.grad(F[2],y,retain_graph=True)[0]-torch.autograd.grad(F[1],z,retain_graph=True)[0]
roty=torch.autograd.grad(F[0],z,retain_graph=True)[0]-torch.autograd.grad(F[2],x,retain_graph=True)[0]
rotz=torch.autograd.grad(F[1],x,retain_graph=True)[0]-torch.autograd.grad(F[0],y,retain_graph=True)[0]

rot=torch.stack([rotx,roty,rotz])
print("Rotacional de F(x, y, z):", rot)
````

- Campo vetorial F = [F1, F2, F3] = [yz, xz, xy]

### 3. Cálculo do Rotacional
O rotacional é calculado como:
```
∇ × F = (∂F₃/∂y - ∂F₂/∂z, ∂F₁/∂z - ∂F₃/∂x, ∂F₂/∂x - ∂F₁/∂y)
```

```python
rotz = torch.autograd.grad(F[1], x, retain_graph=True)[0] - torch.autograd.grad(F[0], y, retain_graph=True)[0]
rot = torch.stack([rotx, roty, rotz])
```
- Cada componente do rotacional é calculada usando derivadas parciais
- `retain_graph=True` permite reutilizar o grafo computacional

## Resultados e Verificação

### Cálculo Analítico
Para F = [yz, xz, xy], o rotacional teórico é:
```

```

Dados: Modulo de Young(E)=200Gpa e ni=0.3(módulo de cisalhamento)

````python
import torch
#DADOS
E=210e9
ni=0.3
#PARÂMETROS DE LAMÉ
Lambda=(E*ni)/((1+ni)*(1-2*ni))
mi=E/(2*(1+ni))
#u(x,y)=ax, v(x,y)=by
a=0.01
b=0.02
#MALHA DE PONTOS
x=torch.linspace(0,1,10, requires_grad=True)
y=torch.linspace(0,1,10, requires_grad=True)
X,Y=torch.meshgrid(x,y,indexing='ij')
#CAMPOS DE DESLOCAMENTO
u=a*X
v=b*Y
#DERIVADAS
dudx=torch.autograd.grad(u.sum(),x,create_graph=True)[0][0]
dvdy=torch.autograd.grad(v.sum(),y,create_graph=True)[0][0]
#TENSOR DE DEFORMAÇÔES EPSILON
epsxx=dudx
epsyy=dvdy
TReps=epsxx+epsyy
#TENSÔES
sigmaxx=Lambda*TReps+2*mi*epsxx
sigmayy=Lambda*TReps+2*mi*epsyy

print(f"ε_xx={epsxx.item():.4e}, ε_yy = {epsyy.item():.4e}")
print(f"σ_xx={sigmaxx.item():.4e} Pa, σ_yy = {sigmayy.item():.4e} Pa")
#print("Lambda",Lambda)
#print("Mi",mi)
````

##Regressão Linear
Execício adaptado da lista https://github.com/cmpmech/deep-learning-in-
computational-mechanics, Capítulo 2 ex.4


````python
import torch
import matplotlib.pyplot as plt
import numpy as np

# 1. Geração de dados com ruído
torch.manual_seed(7654)
xTreino= torch.randn(80)
yTreino=2* xTreino+ 3 + torch.randn(80) # Corrected variable name

xTeste = torch.randn(20)
yTeste = 2*xTeste + 3 + torch.randn(20)

# 2. Modelo de Regressão Linear com PyTorch
class LinearRegression(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.peso = torch.nn.Parameter(torch.tensor(0.0))
        self.vies = torch.nn.Parameter(torch.tensor(0.0))

    def forward(self, x):
        return self.peso * x + self.vies

# 3. Inicialização do modeloo e otimizador
modelo = LinearRegression()
otimizador = torch.optim.SGD(modelo.parameters(), lr=0.1)
custo = torch.nn.MSELoss()

# 4. Treinamento
epochs = 100
for epoch in range(epochs):
    # Forward pass
    yPrev = modelo(xTreino)
    PerdaTreino = custo(yPrev, yTreino)

    # Backward pass
    otimizador.zero_grad()
    PerdaTreino.backward()
    otimizador.step()

    # Avaliação
    if epoch % 10 == 0:
        PerdaTeste = custo(modelo(xTeste), yTeste)
        print(f"Época {epoch}/{epochs} - Custo de Treino: {PerdaTreino.item():.4e} - Custo de Teste: {PerdaTeste.item():.4e}")

# 5. Visualização
x_plot = torch.linspace(xTeste.min(), xTeste.max(), 100)
y_plot = modelo(x_plot).detach()

plt.figure(figsize=(10, 4))
plt.scatter(xTreino.numpy(), yTreino.numpy(), color='k', label='Dados de Treinamento')
plt.scatter(xTeste.numpy(), yTeste.numpy(), color='r', label='Dados Teste')
plt.plot(x_plot.numpy(), y_plot.numpy(), color='b', label='Previsão')
plt.grid(True)
plt.legend()
plt.title("Regressão Linear via PyTorch")
plt.show()

# 6. Parâmetros aprendidos
print("\nParâmetros Aprendidos:")
print(f"Viés (intercepto): {modelo.vies.item():.3f}")
print(f"Peso (inclinação): {modelo.peso.item():.3f}")

# 7. Comparação com solução analítica (via NumPy)
x_np = xTreino.numpy()
y_np = yTreino.numpy()
X = np.vstack([np.ones_like(x_np), x_np]).T
theta = np.linalg.inv(X.T @ X) @ X.T @ y_np
print("\nSolução da Equação Normal (Viés, Peso):", theta)
````

##Regressão não linear

````python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

# 1. Gerar dados (x, y)
x = torch.linspace(-2 * torch.pi, 2 * torch.pi, 100).view(-1, 1)
y = torch.sin(x) + 0.1 * torch.randn_like(x)  # seno com ruído

# 2. Definir modelo (rede neural simples com camada oculta)
modelo = nn.Sequential(
    nn.Linear(1, 100),  # entrada -> camada oculta com 20 neurônios
    nn.Tanh(),         # ativação não linear
    nn.Linear(100, 1)   # camada oculta -> saída
)

# 3. Definir otimizador e função de perda
otim = torch.optim.Adam(modelo.parameters(), lr=0.01)  # otimizador
criterio = nn.MSELoss()  # função de erro: erro quadrático médio

# 4. Treinar a rede
for epoca in range(1000):  # repetir 1000 vezes (épocas)
    pred = modelo(x)              # prever y com base em x
    loss = criterio(pred, y)      # calcular erro (diferença entre y previsto e real)

    otim.zero_grad()  # zera os gradientes anteriores
    loss.backward()   # computa gradientes automaticamente
    otim.step()       # atualiza os pesos da rede

# 5. Visualizar resultado
plt.figure(figsize=(10, 2))
plt.plot(x.numpy(), y.numpy(), 'o', label='dados')  # pontos reais com ruído
plt.plot(x.numpy(), modelo(x).detach().numpy(), 'r-', label='modelo')  # curva ajustada
plt.grid(True)
plt.legend()
plt.show()
````

````python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt

#Hiperparâmetros que podem ser alterados
input_size = 1
hidden_size = 100
output_size = 1
learning_rate = 0.01
num_epochs = 1000

#Dados de exemplo
x = torch.unsqueeze(torch.linspace(-2, 2, 100), dim=1)
y = torch.sin(x) + 0.1 * torch.randn_like(x)

#Modelo
model = nn.Sequential(
    nn.Linear(input_size, hidden_size),
    nn.ReLU(),
    nn.Linear(hidden_size, output_size)
)

# Função de custo e otimizador
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=learning_rate)

#Treinamento
for epoch in range(num_epochs):
    pred = model(x)
    loss = criterion(pred, y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch+1) % 100 == 0:
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")

#Visualização
predicted = model(x).detach()  # .detach() remove do grafo computacional

plt.figure(figsize=(8, 2))
plt.scatter(x.numpy(), y.numpy(), label='Dados reais', alpha=0.5)
plt.plot(x.numpy(), predicted.numpy(), label='Previsão da rede', color='red')
plt.legend()
plt.title("Ajuste da rede neural aos dados")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
````

## Explicação do Código

### 1. Configuração dos Hiperparâmetros
```python
input_size = 1    # Dimensão de entrada (apenas x)
hidden_size = 10  # Neurônios na camada oculta
output_size = 1   # Dimensão de saída (apenas y)
learning_rate = 0.01  # Taxa de aprendizado
num_epochs = 1000    # Número de iterações de treinamento
```

### 2. Geração de Dados Sintéticos
```python
x = torch.unsqueeze(torch.linspace(-2, 2, 100), dim=1)  # 100 pontos entre -2 e 2
y = torch.sin(x) + 0.1 * torch.randn_like(x)  # Função seno com ruído gaussiano
```
- **Objetivo**: Aprender a aproximar a função seno a partir de dados ruidosos
- **Ruído**: Adicionado para simular dados reais (10% da amplitude)

### 3. Arquitetura da Rede Neural
```python
model = nn.Sequential(
    nn.Linear(input_size, hidden_size),  # Camada linear de entrada
    nn.ReLU(),                          # Função de ativação não-linear
    nn.Linear(hidden_size, output_size) # Camada linear de saída
)
```
- **Estrutura**: MLP (Multilayer Perceptron) com 1 camada oculta
- **Ativação**: ReLU (Rectified Linear Unit) introduz não-linearidade

### 4. Configuração do Treinamento
```python
criterion = nn.MSELoss()  # Erro Quadrático Médio
optimizer = optim.SGD(model.parameters(), lr=learning_rate)  # Gradiente Descendente
```

### 5. Loop de Treinamento
```python
for epoch in range(num_epochs):
    pred = model(x)          # Forward pass
    loss = criterion(pred, y)  # Cálculo do erro
    
    optimizer.zero_grad()    # Zera gradientes
    loss.backward()          # Backpropagation
    optimizer.step()         # Atualiza pesos
    
    if (epoch+1) % 100 == 0:
        print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")
```
- **Monitoramento**: Exibe a perda a cada 100 epochs

### 6. Visualização dos Resultados
```python
predicted = model(x).detach()  # Remove do grafo computacional para plotagem

plt.scatter(x.numpy(), y.numpy(), label='Dados reais', alpha=0.5)
plt.plot(x.numpy(), predicted.numpy(), label='Previsão', color='red')
```
- **Gráfico**: Compara dados originais com a previsão da rede
